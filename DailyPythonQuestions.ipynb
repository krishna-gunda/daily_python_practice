Problem Statement: Write a function count_vowels(s) that takes a string and returns the count of vowels
 (a, e, i, o, u — case-insensitive) in it.

** Title — "Second Largest Element"

Problem Statement: Write a function second_largest(lst) that takes a list of integers and returns the second largest unique element. If no second largest exists, return None.

Example: [4, 1, 7, 7, 3] → 4

Difficulty Level: Easy

Expected Concepts Used: Lists, Sorting, Sets**

```python
def second_largest(lis):  #[2,4,1,6,7]
  first=0
  second=0
  for i in lis:
    if i>first:
      second=first
      first=i
    elif i>second and i!=first:
      second=i

  return second

second_largest([10, 8, 5])

```

Reverse Words in a Sentence"

Problem Statement: Write a function reverse_words(sentence) that takes a string and returns the sentence with the order of words reversed — without using any built-in reverse functions.

Example: "I love Python" → "Python love I"

Difficulty Level: Easy

Expected Concepts Used: Strings, Loops, Lists

```python
def reverse_words(sentence):
  words=sentence.split()
  s=" ".join(words[::-1])
  return s

reverse_words('i love python')
```

Title — "Lambda-Based Student Sorter"

Problem Statement: You have a list of tuples representing students: (name, age, marks). Write a one-liner using sorted() and a lambda function to sort students — first by marks in descending order, and if marks are equal, by name in ascending alphabetical order.

Example: [("Alice", 20, 88), ("Bob", 21, 95), ("Charlie", 19, 88)]

Difficulty Level: Medium

Expected Concepts Used: Lambda Functions, Tuples, Sorting

```python
a = [("Alice", 20, 88), ("Charlie", 19, 88), ("Bob", 21, 95)]

for i in range(len(a)):
    for j in range(i + 1, len(a)):

        # Sort by marks descending
        if a[i][2] < a[j][2]:
            a[i], a[j] = a[j], a[i]

        # If marks equal, sort by name ascending
        elif a[i][2] == a[j][2]:
            if a[i][0] > a[j][0]:
                a[i], a[j] = a[j], a[i]

print(a)
```

Title — "Inventory Manager"

Problem Statement: Write a function update_inventory(inventory, sales) where inventory is a dictionary of {item: quantity} and sales is a list of (item, sold_quantity) tuples. The function should:

Deduct sold quantities from inventory
Raise a custom exception InsufficientStockError if sold quantity exceeds available stock
Return the updated inventory

Example:
inventory = {"apple": 10, "banana": 5}
sales = [("apple", 4), ("banana", 7)]
→ Should raise InsufficientStockError for banana
Difficulty Level: Interview Level

Expected Concepts Used: Dictionaries, Exception Handling, Loops, Tuples

```python
inventory = {"apple": 10, "banana": 5}
sales = [("apple", 4), ("banana", 7)]
def update_inventory(inventory,sales):
  for item, qty in sales:
      if item in inventory:
          if qty > inventory[item]:
              yield (f'The inventory is out of stock for {item}')
          else:
              yield (f'The inventory is in stock for {item}')
```

```python
inventory = {"apple": 10, "banana": 5}
sales = [("apple", 4), ("banana", 7)]
for i in update_inventory(inventory,sales):
  print(i)
```
